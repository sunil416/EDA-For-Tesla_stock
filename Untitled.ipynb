```python
dataframe_tesla["5 Day Rolling Close"]= dataframe_tesla["Close"].rolling(5, min_periods=1).mean()
dataframe_tesla["10 Day Rolling Close"]= dataframe_tesla["Close"].rolling(10, min_periods=1).mean()
dataframe_tesla["15 Day Rolling Close"]= dataframe_tesla["Close"].rolling(15, min_periods=1).mean()
dataframe_tesla["30 Day Rolling Close"]= dataframe_tesla["Close"].rolling(30, min_periods=1).mean()
```

```python
dataframe_tesla[["Close","5 Day Rolling Close","10 Day Rolling Close","15 Day Rolling Close","30 Day Rolling Close"]].plot(xlim=["01-01-2020","16-06-2022"], figsize=(15,7))
```

```python
dataframe_tesla["EMA_0.1"]= dataframe_tesla["Close"].ewm(alpha=0.1, adjust=False).mean()
dataframe_tesla["EMA_0.3"]= dataframe_tesla["Close"].ewm(alpha=0.3, adjust=False).mean()
```

```python
dataframe_tesla[["Close","EMA_0.1","EMA_0.3"]].plot(xlim=["01-01-2020","16-06-2022"], figsize=(15,7))  ## Exponential Moving Average 
```

```python
import statsmodels.api as sms
```

```python
from statsmodels.tsa.stattools import adfuller
```

```python
def adfularTest(series):
    output=adfuller(series)
    if output[1]>0.5:
        print(f"Data is not stationary null hypthisis is accepted p= {output[1]}")
    else:
        print(f"Data is stationary, alternate Null hyposthis is accepted p= {output[1]}")
```

```python
adfularTest(dataframe_tesla['Close'])
```

```python
# Calcaulting the 1st diffrence 
dataframe_tesla["First Diff"]= dataframe_tesla['Close']-dataframe_tesla['Close'].shift(1)
```

```python
adfularTest(dataframe_tesla["First Diff"].dropna())
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
plot_acf(dataframe_tesla["First Diff"].dropna())  ### q=7  , Moving Average is performed
```

```python
plot_pacf(dataframe_tesla["First Diff"].dropna()) ## p=7 , Auto Regression
```

```python
import xgboost as xgb
```

```python
from xgboost import plot_importance, plot_tree
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
reg = xgb.XGBRegressor(n_estimators=1000)

dataframe_tesla=dataframe_tesla.reset_index()
```

```python
dataframe_tesla["Day"]=dataframe_tesla["Date"].dt.day
dataframe_tesla["Month"]=dataframe_tesla["Date"].dt.month
dataframe_tesla["Year"]=dataframe_tesla["Date"].dt.year
dataframe_tesla.drop("Date", axis=1, inplace=True)
```

```python
from datetime import datetime
```

```python
train_data = dataframe_tesla[:-100]
test_data= dataframe_tesla[dataframe_tesla.shape[1]-100:]
```

```python
y_train = train_data["Close"]
X_train = train_data.drop("Close", axis=1)
y_test=test_data["Close"]
X_test=test_data.drop("Close", axis=1)
```

```python
dataframe_tesla.info()
```

```python
reg.fit(X_train, y_train,
        eval_set=[(X_train, y_train), (X_test, y_test)],
        early_stopping_rounds=50,
       verbose=False)
```

```python
plot_importance(reg, height=0.9)
```
